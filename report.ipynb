```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
import arabic_reshaper
from bidi.algorithm import get_display

# تابع برای پردازش متن فارسی
def reshape_text(text):
    reshaped = arabic_reshaper.reshape(text)  # اصلاح شکل حروف
    return get_display(reshaped)  # تبدیل به ترتیب صحیح

# تابع برای تبدیل عدد انگلیسی به فارسی
def to_persian_number(text):
    persian_digits = {'0': '۰', '1': '۱', '2': '۲', '3': '۳', '4': '۴', '5': '۵', '6': '۶', '7': '۷', '8': '۸', '9': '۹'}
    return ''.join([persian_digits.get(i, i) for i in text])

# مسیر فونت فارسی
font_path_fa = "/home/aliebrahimian001/Desktop/github/Autism-Diagnosis/b-nazanin/BNazanin.ttf"  # مسیر دقیق فایل فونت فارسی
font_prop_fa = fm.FontProperties(fname=font_path_fa)

# مسیر فونت انگلیسی
font_path_en = "/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf"  # مسیر فونت انگلیسی
font_prop_en = fm.FontProperties(fname=font_path_en)

# داده‌ها
models = ['VGG19', 'VGG16', 'EfficientNetB0']
learning_rate_001 = [74.00, 76.22, 87.37]
learning_rate_0001 = [50.74, 52.74, 88.33]
learning_rate_00001 = [84.66, 80.05, 87.9]

# تنظیمات نمودار
fig, ax = plt.subplots(figsize=(10, 6))

# عرض بارها
bar_width = 0.2
index = np.arange(len(models))

# رسم نمودار
bar1 = ax.bar(index, learning_rate_001, bar_width, label=reshape_text('نرخ یادگیری 0.01'), color='#ff6347')
bar2 = ax.bar(index + bar_width, learning_rate_0001, bar_width, label=reshape_text('نرخ یادگیری 0.001'), color='#32cd32')
bar3 = ax.bar(index + 2 * bar_width, learning_rate_00001, bar_width, label=reshape_text('نرخ یادگیری 0.0001'), color='#1e90ff')

# اضافه کردن مقادیر داخل هر ستون با اعداد فارسی
for i, value in enumerate(learning_rate_001):
    ax.text(i, value + 1, to_persian_number(f'{value:.2f}%'), ha='center', va='bottom', fontproperties=font_prop_en)
for i, value in enumerate(learning_rate_0001):
    ax.text(i + bar_width, value + 1, to_persian_number(f'{value:.2f}%'), ha='center', va='bottom', fontproperties=font_prop_en)
for i, value in enumerate(learning_rate_00001):
    ax.text(i + 2 * bar_width, value + 1, to_persian_number(f'{value:.2f}%'), ha='center', va='bottom', fontproperties=font_prop_en)

# تنظیمات نمودار
ax.set_xlabel(reshape_text('مدل‌ها'), fontproperties=font_prop_fa)
ax.set_ylabel(reshape_text('دقت (%)'), fontproperties=font_prop_fa)
ax.set_title(reshape_text('مقایسه دقت مدل‌ها با نرخ‌های یادگیری مختلف'), fontsize=20, fontweight="bold", fontproperties=font_prop_fa)

# تنظیمات برای مدل‌ها به زبان انگلیسی
ax.set_xticks(index + bar_width)
ax.set_xticklabels(models, fontproperties=font_prop_en)

# اضافه کردن لگند
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), prop=font_prop_fa)

# نمایش نمودار
plt.tight_layout()
plt.savefig('img/learning_rate_comparison.webp', format='webp')
plt.show()
```
